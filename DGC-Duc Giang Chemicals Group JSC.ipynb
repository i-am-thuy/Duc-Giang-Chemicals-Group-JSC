##DGC – CÔNG TY CỔ PHẦN TẬP ĐOÀN HÓA CHẤT ĐỨC GIANG 

+ DGC tiền thân là Công ty Hoá chất Đức Giang, là một doanh nghiệp nhà nước trực thuộc Tổng cục hóa chất Việt Nam được thành lập từ năm 1963. Tháng 3/2004, Công ty thực hiện chuyển đổi mô hình và chính thức hoạt động dưới hình thức Công ty cổ phần với tên gọi là Công ty Cổ phần Bột giặt và Hóa chất Đức Giang. Năm 2009, DGC thành lập Công ty con - Công ty cổ phần Hóa chất Đức Giang Lào Cai. Tháng 4/2019: Chuyển đổi mô hình tổ chức và hoạt động sang mô hình Tập đoàn và đổi tên thành Công ty cổ phần Tập đoàn Hóa chất Đức Giang như hiện nay.
+	Lĩnh vực kinh doanh chính: Các sản phẩm kinh doanh mang lại nguồn doanh thu chính cho Công ty là phốt pho vàng và bột giặt, chất tẩy rửa. Ngoài ra, DGC cũng sản xuất và kinh doanh thương mại nhiều sản phẩm hóa chất cơ bản khác. Công ty còn triển khai mở rộng sản xuất sang lĩnh vực hóa chất nông nghiệp, phân bón.
+	DGC là nhà sản xuất và xuất khẩu photpho vàng hàng đầu và các sản phẩm liên quan như axit photphoric, phân lân và phụ gia thức ăn chăn nuôi. Phần lớn các sản phẩm của DGC đều phục vụ cho thị trường xuất khẩu, thị trường xuất khẩu chính là Ấn Độ, sau đó là các nước châu Á (Nhật Bản, Hàn Quốc, Đài Loan), bên cạnh đó DGC cũng xuất khẩu sang thị trường Mỹ và EU.

##KẾT QUẢ KINH DOANH QUÝ I/2022:

+	Q1/2022, DGC ghi nhận lợi nhuận ròng đạt 1335 tỷ (+370.4% QOQ), xấp xĩ gấp 5 lần so với Q1/2021 và doanh thu đạt 3634 tỷ (+86.5% QOQ) nhờ giá bán các mặt hàng, đặc biệt là Photpho vàng đều tăng vọt do Trung Quốc cắt giảm các sản phẩm công nghiệp nặng và đóng cửa để thực hiện chính sách Zero-Covid. Giá Photpho trên thế giới có sự giằng co nhưng giá bán tiếp tục duy trì đà tăng trong Q1/22, giá bán bình quân 126-127 triệu/ tấn (+120% so với cùng kỳ).

##LUẬN ĐỂM ĐẦU TƯ:

+	Về ngắn hạn, động lực tăng trưởng nhờ:

(1) Hai dây chuyền axitphotphoric (TPA) cấp điện tử mới dự kiến hoạt động 80% công suất trong năm 2022 nhờ nhu cầu của chất bán dẫn và thiết bị 5G tiếp tục tăng cao trong 2022. 

(2) Hưởng lợi ngắn hạn từ xung đột Nga-Ukraine và thiếu hụt nguồn cung photpho từ Trung Quốc.
+	Về dài hạn, động lực tăng trưởng dài hạn của DGC đến từ:

(1)	Dự án Chlor-alkali-vinyl (CAV) – Nghi Sơn dự kiến hoàn thành giai đọan 1 trong Q1/2025, sẽ đưa DGC trở thành doanh nghiệp có nhà máy xút với công suất thương mại số 1 Việt Nam. 

(2)	Mỏ quặng apatit thứ hai của DGC dự kiến sẽ giúp công ty tiết kiệm 20-30% chi phí quặng đầu vào hằng năm từ dự án khai trường 25 - Lào Cai. (DGC có thể tự chủ quặng Apatit trong 6 năm, kể từ năm 2021)


##RỦI RO ĐẦU TƯ:

+	Chung cư Hóa chất Đức Giang đang gặp khó khăn trong quá trình cấp phép chủ trương đầu tư, chưa được triển khai chính thức.
+	Giá photpho vàng đang biến động mạnh theo diễn biến căng thẳng giữa Nga – Ukraine, nếu tình hình xung đột giữa hai nước bình ổn lại và Trung Quốc mở cửa thì giá photpho sẽ giảm.
